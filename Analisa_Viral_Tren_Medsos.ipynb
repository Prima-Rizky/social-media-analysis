**Import library yang diperlukan**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Load dataset

```python
df = pd.read_csv('Viral_Social_Media_Trends.csv')
```

Tampilkan beberapa baris pertama dari dataset

```python
print("Beberapa baris pertama dari dataset:")
print(df.head())
```

Analisis statistik dasar

```python
print("\nStatistik Deskriptif:")
print(df.describe())
```

Cek nilai yang hilang dalam dataset

```python
print("\nJumlah nilai yang hilang per kolom:")
print(df.isnull().sum())
```

Menghitung frekuensi hashtag populer

```python
hashtag_populer = df['Hashtag'].value_counts()
print("\nFrekuensi Hashtag Populer:")
print(hashtag_populer)
```

Visualisasi frekuensi hashtag

```python
plt.figure(figsize=(10, 6))
hashtag_populer.head(10).plot(kind='bar', color='skyblue')
plt.title('Top 10 Hashtag Populer')
plt.xlabel('Hashtag')
plt.ylabel('Jumlah Postingan')
plt.xticks(rotation=45)
plt.show()
```

Menghitung frekuensi platform sosial media yang digunakan

```python
platform_populer = df['Platform'].value_counts()
print("\nFrekuensi Platform Sosial Media:")
print(platform_populer)
```

Visualisasi platform sosial media

```python
plt.figure(figsize=(10, 6))
platform_populer.plot(kind='bar', color='lightcoral')
plt.title('Jumlah Postingan per Platform Sosial Media')
plt.xlabel('Platform')
plt.ylabel('Jumlah Postingan')
plt.xticks(rotation=45)
plt.show()
```

Analisis hubungan antara Views, Likes, Shares, dan Comments

Visualisasi distribusi Views, Likes, Shares, dan Comments

```python
plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
sns.histplot(df['Views'], kde=True, color='blue')
plt.title('Distribusi Views')

plt.subplot(2, 2, 2)
sns.histplot(df['Likes'], kde=True, color='green')
plt.title('Distribusi Likes')

plt.subplot(2, 2, 3)
sns.histplot(df['Shares'], kde=True, color='red')
plt.title('Distribusi Shares')

plt.subplot(2, 2, 4)
sns.histplot(df['Comments'], kde=True, color='purple')
plt.title('Distribusi Comments')

plt.tight_layout()
plt.show()
```

Analisis Korelasi antara Views, Likes, Shares, dan Comments

```python
corr_matrix = df[['Views', 'Likes', 'Shares', 'Comments']].corr()
print("\nMatriks Korelasi antara Views, Likes, Shares, dan Comments:")
print(corr_matrix)
```

Visualisasi matriks korelasi

```python
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Matriks Korelasi')
plt.show()
```

Analisis berdasarkan Region


Melihat jumlah postingan berdasarkan region

```python
region_populer = df['Region'].value_counts()
print("\nJumlah Postingan berdasarkan Region:")
print(region_populer)
```

Visualisasi jumlah postingan berdasarkan region

```python
plt.figure(figsize=(10, 6))
region_populer.plot(kind='bar', color='lightgreen')
plt.title('Jumlah Postingan berdasarkan Region')
plt.xlabel('Region')
plt.ylabel('Jumlah Postingan')
plt.xticks(rotation=45)
plt.show()
```

Analisis Tingkat Keterlibatan

Menghitung jumlah postingan berdasarkan tingkat keterlibatan

```python
engagement_level = df['Engagement_Level'].value_counts()
print("\nJumlah Postingan berdasarkan Tingkat Keterlibatan:")
print(engagement_level)
```

Visualisasi tingkat keterlibatan

```python
plt.figure(figsize=(8, 6))
engagement_level.plot(kind='bar', color='salmon')
plt.title('Jumlah Postingan berdasarkan Tingkat Keterlibatan')
plt.xlabel('Tingkat Keterlibatan')
plt.ylabel('Jumlah Postingan')
plt.xticks(rotation=0)
plt.show()
```
